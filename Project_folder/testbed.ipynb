```python
trunc_desc = description.drop([0, 1, 30, 31, 33, 34, 35], axis=0)
trunc_desc = trunc_desc.reset_index(drop=True)
print(trunc_desc.shape)

def swap_rows(df, row1, row2): #Taken from Statolog.org
    df.iloc[row1], df.iloc[row2] =  df.iloc[row2].copy(), df.iloc[row1].copy()
    return df


trunc_desc = swap_rows(trunc_desc, 25, 27)
trunc_desc = swap_rows(trunc_desc, 26, 28)
trunc_desc["Variable Name"] = trunc_desc["Variable Name"].str.strip() # four of the trunc values from the description have spaces which makes them not the same as the data column names
display(trunc_desc)
```

```python
possible_values = ["1 - Yes | 0 - No", "1 - Male | 2 - Female", "1 - dead | 0 - alive"]
mask = trunc_desc["Values"].isin(possible_values)
Binary_indexes = trunc_desc[mask].index.tolist()
Num_indexes = trunc_desc["Values"].drop(Binary_indexes).index.tolist()
Binary_indexes = Binary_indexes[:-1]
Num_indexes = Num_indexes[:-1]
print(Binary_indexes)
print(Num_indexes)
print(len(Binary_indexes) + len(Num_indexes))
```

### Notes on the various terms:  
Major comorbidity: Another significant medical condition or disease  
node positive primary: cancer has spread to lymph nodes  
synchronous_crlm: more than one primary cancer detected in the liver at the time of initial presentation  
months_to_liver_DFS_progression: DFS stands for Disease Free Survival, and this indicates the time from presentation to survival without any signs or symptoms of the disease  
vital_status_liver_DFS:	this must be wether the liver survived without transplant?  
vital_status_DFS: wether the patient survived while being disease free?  
clinrisk_score: some calculated risk  
De-identify Scout Name: just an anonymizer (not really relevant outside of being an assigned ID with respect to the dataset)  
clinrisk_stratified: says something about the clinical risk strategy used. Might mean that there is a subjective measure in addition to quantifiable data for determining the risk of the patient (like doctors judgment).
